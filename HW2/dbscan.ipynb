```python
# DBScan: https://scikit-learn.org/stable/modules/generated/sklearn.cluster.DBSCAN.html

from sklearn.neighbors import NearestNeighbors
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs, make_moons

import matplotlib.pyplot as plt
import numpy as np
```

```python
X, y = make_moons(n_samples=100)
```

```python
plt.figure(figsize=(9,6))
scatter = plt.scatter(X[:,0], X[:,1])
```

```python
# Classify data using DBSCAN (from sklearn)
# Plot results with corresponding colors
```

```python
clustering = DBSCAN().fit(X)
labels = clustering.labels_

plt.figure(figsize=(9,6))
scatter = plt.scatter(X[:,0], X[:,1], c=labels)
```

```python
np.random.seed(42)
X, _ = make_blobs(n_samples=1000,
                  centers=15,
                  n_features=2,
                  cluster_std=np.random.uniform(low=0.5, high=1.2, size=(15)),
                  random_state=42)
```

```python
plt.figure(figsize=(9,6))
scatter = plt.scatter(X[:,0], X[:,1])
```

```python
# Find best eps value using the Elbow-method and plot results
# https://towardsdatascience.com/machine-learning-clustering-dbscan-determine-the-optimal-value-for-epsilon-eps-python-example-3100091cfbc
```

```python
neighbors = NearestNeighbors(n_neighbors=2)
nbrs = neighbors.fit(X)
distances, indices = nbrs.kneighbors(X)

distances = np.sort(distances, axis=0)
plt.plot(distances[:, 1])
plt.grid()
```

```python
clustering = DBSCAN(eps=0.5, min_samples=6).fit(X)
labels = clustering.labels_
print(set(labels))  #  15   

plt.figure(figsize=(9,6))
scatter = plt.scatter(X[:,0], X[:,1], c=labels)
```
